**Task: Add following columns to master dataframe: <br>**

- Make home team name in lower case
- Make visit team name in upper case
- Replace '/' in link column with '#"
- Extract first 5 characters from link column
- Add flag = 1 if link columns contains value 'CHO' else 0
- Split home team name based on spaces and make new columns for each split value. E.g. 'A B' will have 2 columns with values 'A' and 'B'
- Join back split columns from above and check if it is same as home team name
- Some names in home team name have spaces around them. Find a way to remove them and make a new column. Add a flag if new column and old home team name columns are same

```python
#1.Make home team name in lower case
dataMaster['homeTeamNameLowerCase'] = dataMaster['home_team_name'].str.lower()
dataMaster.head(2)
```

```python
#2.Make visit team name in upper case
dataMaster['visitor_team_name_UpperCase'] = dataMaster['visitor_team_name'].str.upper()
dataMaster.head(2)
```

```python
#3.Replace '/' in link column with '#"
dataMaster['link'] = dataMaster['link'].str.replace('/', '#')
dataMaster.head(2)
```

```python
#4.Extract first 5 characters from link column
dataMaster['linkNew'] = dataMaster['link'].str[0:5]  
dataMaster.head(2)
```

```python
#5.Add flag = 1 if link columns contains value 'CHO' else 0
dataMaster['linkFlag'] = np.where(dataMaster['link'].str.contains('CHO'), 1, 0)
#np.where() return the indices of elements in an array that satisfy a given condition
dataMaster.head(2)
```

```python
#6.Split home team name based on spaces and make new columns for each split value. 
#E.g. 'A B' will have 2 columns with values 'A' and 'B'
dataHomeTeamName = dataMaster['home_team_name'].str.strip().str.split(' ', expand=True).rename(columns = lambda x: "home_team_name_"+str(x+1)).fillna('')
dataHomeTeamName.head()

dataMaster[['homeTeamName1','homeTeamName2','homeTeamName3']] = dataMaster['home_team_name'].str.strip().str.split(' ', expand=True).rename(columns = lambda x: "home_team_name_"+str(x+1)).fillna('')
```

```python
#7.Join back split columns from above and check if it is same as home team name
dataMaster['homeTeamNameJoined'] = dataHomeTeamName.agg(' '.join, axis=1)
dataMaster.head(2)
```

```python
#8.Some names in home team name have spaces around them. Find a way to remove them and make a new column.
#Add a flag if new column and old home team name columns are same
dataMaster['homeTeamNameClean'] = dataMaster['home_team_name'].str.strip() # remeove the space
dataMaster['homeTeamNameCleanMatch'] = dataMaster['home_team_name'] == dataMaster['homeTeamNameClean'] #add a flag

#Since Q17 need column visitor_team_name, we will clean it here. 
dataMaster['visitorTeamNameClean'] = dataMaster['visitor_team_name'].str.strip().replace(r'[^\w ]', '', regex=True)
```

**Task: Group by home team name and season, find average of attendance. Convert it to wide format. Reconvert it to long format. Show multiple ways of doing that. Use method chaining while writing code**

```python
dataMaster['attendance'] = pd.to_numeric(dataMaster['attendance'].str.replace(',', ''), errors='coerce') #convert the data type
```

```python
dataMaster_groupBy = dataMaster.groupby(['homeTeamNameClean', 'season'])['attendance'].mean()
dataMaster_groupBy
```

```python
#Using unstack() to create wide format, stack() to create long format

wideFormat1 = dataMaster_groupBy.unstack(level =1)
longFormat1 = wideFormat1.stack()

#Using pivot_table() to create wide format, melt() to create long format 
wideFormat2 = dataMaster_groupBy.to_frame()
wideFormat2 = wideFormat2.pivot_table(index='homeTeamNameClean', columns='season', values='attendance')
# wideFormat2 

wideFormat2  = wideFormat2.reset_index()
longFormat2 = pd.melt(wideFormat2 , id_vars='homeTeamNameClean', value_vars='season', value_name='attendance' )
# longFormat2
```

**Task: Fill na values with 'missing' or 0 on the basis of field type of column**

```python
columnDataTypes = dataMaster.apply(lambda x: x.dtype)
for column, data_type in columnDataTypes.iteritems():
    if data_type == 'int64':
        dataMaster[column].fillna(0, inplace=True)
    elif data_type == 'float64':
        dataMaster[column].fillna(0, inplace=True)
    elif data_type == 'int32':
        dataMaster[column].fillna(0, inplace=True)       
    elif data_type == 'object':
        dataMaster[column].fillna('missing', inplace=True)
```

**Task: Create variable having average score so far for each team in the existing dataset**

```python
# Sort the DataFrame by date
dataMaster = dataMaster.sort_values('timestamp')
dataMaster['home_cumulative_score'] = dataMaster.groupby('homeTeamNameClean')['home$score'].cumsum()
dataMaster['home_cumulative_count'] = dataMaster.groupby('homeTeamNameClean').cumcount() + 1
dataMaster['home_average_score'] = dataMaster['home_cumulative_score'] / dataMaster['home_cumulative_count']
dataMaster['visitor_cumulative_score'] = dataMaster.groupby('visitorTeamNameClean')['visitor$score'].cumsum()
dataMaster['visitor_cumulative_count'] = dataMaster.groupby('visitorTeamNameClean').cumcount() + 1
dataMaster['visitor_average_score'] = dataMaster['visitor_cumulative_score'] / dataMaster['visitor_cumulative_count']

dataMaster[['timestamp','homeTeamNameClean','visitorTeamNameClean','home_average_score','visitor_average_score']].head(10)
```
